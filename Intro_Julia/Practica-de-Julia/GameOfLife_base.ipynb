```julia
#Salto temporal
function turno(T)
    NT = copy(T)
    for i in 1:N ,j in 1:N
        aux = 0
        for l in [i-1,i,i+1]
        #print(l)
            for m in [j-1,j,j+1]
                #println("Fila $l columuna $m")
                if (l==i)&(m==j)
                    continue
                end
                try
                    #println("Posicion $i ,$j"Ci[l,m])
                    aux += T[l,m]
                catch
                    continue
                end
            end
        end
        vec = aux
        #print(vec)
        if T[i,j] == 0
            if vec == 3     #Nacimiento
                NT[i,j] = 1
            end
        else
            if (vec < 2)||(vec > 3)     #Muerte
                NT[i,j] = 0
            end
        end
    end
    return NT
end
#Visualización
function Table(Tablero,dim,Color)
    Mesa = rand(Color,dim,dim)
    for i in 1:N ,j in 1:N
        (Tablero[i,j] == 0) ? Mesa[i,j] = Color[1] : Mesa[i,j] = Color[2]
    end
    return Mesa
end
```

```julia
# Game of life CI
Estados = [colorant"black",colorant"royalblue1"] #distinguishable_colors(2)
#println(typeof(Estados))
N = 25 #Casillas
Ci = rand(0:1,N,N)
```

```julia
Table(Ci,N,Estados)
```

```julia
turno(Ci)
```

```julia
Table(turno(Ci),N,Estados)
```
